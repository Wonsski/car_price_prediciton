```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Loading data

```
DATA_PATH = 'data/car.h5'
```

```
df = pd.read_hdf(DATA_PATH)
df.shape
```

Example data

```
df.sample(5)
```

# Data visualization

```
df.price_value.hist(bins=100)
```

```
def group_bar_plot(groupby,argument='price_value',funkcje=[np.mean,np.median,np.size],sort='mean',top=50):
  return (
      df
      .groupby(groupby)[argument]
      .agg(funkcje)
      .sort_values(by=sort,ascending=False)
      .head(top)
  ).plot(kind='bar',figsize=(15,5),subplots=True)
```

```
group_bar_plot('param_color');
```

```
group_bar_plot('param_year',sort='size');
```

```
group_bar_plot('param_przebieg',sort='size');
```
